```
import pandas as pd
import numpy as np
import random as rd
```

```
td = pd.read_csv("/content/drive/My Drive/Concatenated Files/FinalDataSet.csv")
```

```
td['CO']=td['CO'].apply(lambda x: np.random.uniform(min(td['CO']),max(td['CO'])) if (np.isnan(x)) else x)
```

```
td['NO2']=td['NO2'].apply(lambda x: np.random.choice([x for x in range(int(min(td['NO2'])),int(max(td['NO2'])))]) if (np.isnan(x)) else x)
```

```
td['O3']=td['O3'].apply(lambda x: np.random.uniform(0,0.09) if (np.isnan(x)) else x)
```

```
td['Pb']=td['Pb'].apply(lambda x: np.random.uniform(0,0.09) if (np.isnan(x)) else x)
```

```
td['PM2.5']=td['PM2.5'].apply(lambda x: np.random.uniform(min(td['PM2.5']),max(td['PM2.5'])) if (np.isnan(x)) else x)
```

```
td['PM10']=td['PM10'].apply(lambda x: np.random.choice([x for x in range(int(min(td['PM10'])),int(max(td['PM10'])))]) if (np.isnan(x)) else x)
```

```
td['SO2']=td['SO2'].apply(lambda x: np.random.choice([x for x in range(int(min(td['SO2'])),int(max(td['SO2'])))]) if (np.isnan(x)) else x)
```

```
td['Precip']=td['Precip'].apply(lambda x: np.random.uniform(min(td['Precip']),max(td['Precip'])) if (np.isnan(x)) else x)
```

```
td['AirMax']=td['AirMax'].apply(lambda x: np.random.choice([x for x in range(int(min(td['AirMax'])),int(max(td['AirMax'])))]) if (np.isnan(x)) else x)
```

```
td['AirMin']=td['AirMin'].apply(lambda x: np.random.choice([x for x in range(int(min(td['AirMin'])),int(max(td['AirMin'])))]) if (np.isnan(x)) else x)
```

```
td['Speed']=td['Speed'].apply(lambda x: np.random.choice([x for x in range(int(min(td['Speed'])),int(max(td['Speed'])))]) if (np.isnan(x)) else x)
```

```
td['Solar']=td['Solar'].apply(lambda x: np.random.choice([x for x in range(int(min(td['Solar'])),int(max(td['Solar'])))]) if (np.isnan(x)) else x)
```

```
td['RHMax']=td['RHMax'].apply(lambda x: np.random.uniform(min(td['RHMax']),max(td['RHMax'])) if (np.isnan(x)) else x)
```

```
td['RHMin']=td['RHMin'].apply(lambda x: np.random.uniform(min(td['RHMin']),max(td['RHMin'])) if (np.isnan(x)) else x)
```
