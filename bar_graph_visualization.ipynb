```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
from statistics import mean
from itertools import groupby
import matplotlib.pyplot as plt

with open("/Users/vihitakesiraju/Documents/Training_Data-v3.csv") as csv_reader:
    data = csv.reader(csv_reader)
    spring = []
    summer = []
    fall = []
    winter = []
    for index, row in enumerate(data):
        if index != 0:
            if row[-2] == '0':
                spring.append(row[-1])
            if row[-2] == '1':
                summer.append(row[-1])
            if row[-2] == '2':
                fall.append(row[-1])
            if row[-2] == '3':
                winter.append(row[-1])
                
spring = list(map(int, spring))
summer = list(map(int, summer))
fall = list(map(int, fall))
winter = list(map(int, winter))


good = [len([i for i in spring if i >= 50 ]), len([i for i in summer if i >= 50 ]),len([i for i in fall if i >= 50 ]),len([i for i in winter if i >= 50 ]),]
moderate = [len([i for i in spring if 51 <= i <= 100 ]), len([i for i in summer if 51 <= i <= 100 ]), len([i for i in fall if 51 <= i <= 100 ]), len([i for i in winter if 51 <= i <= 100 ])]
unhealthy = [len([i for i in spring if 151 <= i <= 200 ]), len([i for i in summer if 151 <= i <= 200 ]), len([i for i in fall if 151 <= i <= 200 ]), len([i for i in winter if 151 <= i <= 200 ])]
unhealthy_sensitive = [len([i for i in spring if 101 <= i <= 150 ]), len([i for i in summer if 101 <= i <= 150 ]), len([i for i in fall if 101 <= i <= 150 ]), len([i for i in winter if 101 <= i <= 150 ])]
very_unhealthy = [len([i for i in spring if 201 <= i <= 300 ]), len([i for i in summer if 201 <= i <= 300 ]), len([i for i in fall if 201 <= i <= 300 ]), len([i for i in winter if 201 <= i <= 300 ])]
hazardous = [len([i for i in spring if i >= 301 ]), len([i for i in summer if i >= 301 ]), len([i for i in fall if i >= 301 ]), len([i for i in winter if i >= 301 ])]

print(good)
print(moderate)
print(unhealthy)
print(unhealthy_sensitive)
season = ['spring', 'summer', 'fall', 'winter']
ind = [x for x, _ in enumerate(season)]


plt.bar(ind, np.add(np.add(np.add(np.add(np.add(moderate,good),unhealthy_sensitive),unhealthy),very_unhealthy),hazardous), width=0.3, label='hazardous', color='maroon')
plt.bar(ind, np.add(np.add(np.add(np.add(moderate,good),unhealthy_sensitive),unhealthy),very_unhealthy), width=0.3, label='very_unhealthy', color='purple')
plt.bar(ind, np.add(np.add(np.add(moderate,good),unhealthy_sensitive),unhealthy), width=0.3, label='unhealthy', color='red')
plt.bar(ind, np.add(np.add(moderate,good),unhealthy_sensitive), width=0.3, label='unhealthy_sensitive', color='orange')
plt.bar(ind, np.add(moderate,good), width=0.3, label='moderate', color='yellow')
plt.bar(ind, good, width=0.3, label='good', color='green')

plt.xticks(ind, season)
plt.ylabel("Count Data with AQI values in specific range ")
plt.xlabel("Season")
plt.rcParams["figure.figsize"] = (10,10)
plt.legend(loc="upper center")
plt.title("Bar Graph")
plt.show()
```
