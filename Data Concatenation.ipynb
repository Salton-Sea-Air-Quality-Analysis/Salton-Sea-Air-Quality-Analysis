```
import pandas as pd
```

```
PM10path = list()
PM25path = list()
SO2path = list()
PBpath = list()
COpath = list()
NO2path = list()
O3path = list()
```

```
for i in range(26):
  year = 1994 + i
  PM10path.append("/content/drive/My Drive/PM10/PM10-" + str(year) + ".csv")
  if year>1998:
    PM25path.append("/content/drive/My Drive/PM2.5/PM2.5-" + str(year) + ".csv")
  SO2path.append("/content/drive/My Drive/SO2/SO2-" + str(year) + ".csv")
  PBpath.append("/content/drive/My Drive/PB/PB-" + str(year) + ".csv")
  COpath.append("/content/drive/My Drive/CO/CO-" + str(year) + ".csv")
  NO2path.append("/content/drive/My Drive/NO2/NO2-" + str(year) + ".csv")
  O3path.append("/content/drive/My Drive/O3/O3-" + str(year) + ".csv")
```

```
#PM 10 Concatenation
PM10final = pd.DataFrame()
for i in PM10path:
  PM10 = pd.read_csv(i)
  PM10 = PM10[['Date','Daily Mean PM10 Concentration']]
  PM10final = pd.concat([PM10final,PM10],join="outer")
```

```
PM10final.to_csv("/content/drive/My Drive/PM10final.csv",index=False)
```

```
#PM25 Concatenation
PM25final = pd.DataFrame()
for i in PM25path:
  PM25 = pd.read_csv(i)
  PM25 = PM25[['Date','Daily Mean PM2.5 Concentration']]
  PM25final = pd.concat([PM25final,PM25],join="outer")
```
