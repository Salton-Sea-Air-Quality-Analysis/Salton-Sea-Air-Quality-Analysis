```python
import csv
import pandas as pd
from statistics import mean
from itertools import groupby
import matplotlib.pyplot as plt


with open('farhaan.csv') as csv_reader:
    data = csv.reader(csv_reader)
    dates = []
    aqi = []
    for index, row in enumerate(data):
        if index != 0:
            dates.append(row[0][-4:])
            aqi.append(row[-1])
    ctr = 0
    dates_group = list({int(k): len(list(v))} for k, v in groupby(dates))
    for index, i in enumerate(dates_group):
        for k, v in i.items():
            ctr += v + 1
            list_mean = mean(list(map(int, aqi[:ctr])))
            dates_group[index][k] = list_mean
            
    new_list = []
    for i in dates_group:
        for k,v in i.items():
            new_dict = {}
            new_dict['year'] = k
            new_dict['aqi'] = v
            new_list.append(new_dict)
    new_list = sorted(new_list, key = lambda i: i['year'])


    mean_year, mean_aqi = [], []
    for i in new_list:
        mean_year.append(i['year'])
        mean_aqi.append(i['aqi'])
        
    import matplotlib.pyplot as plt
    y = mean_aqi
    x1 = mean_year
    fig = plt.figure()
    ax = fig.add_axes([0,0,3,3])
    l1 = ax.plot(x1,y,'go-') 
    ax.set_xlabel('year')
    ax.set_ylabel('AQI')
    ax.set_title("AQI (yearly mean) vs YEAR")
    plt.show()
```
